```python
import glob
from tabula import read_pdf
import pandas as pd
```

```python
# Make a list of all the downloaded files
file_list = sorted(glob.glob('/Users/santhosh_ram/Desktop/Tea/*.pdf'))
# Specify the area in the pdf, inside which our table of interest lies.
top = 217.59
left = 134.31
height = 285.87
width = 324.61
```

```python
#try reading a sample file
ds = read_pdf(file_list[5],pages = 1 ,area = (top,left,height+top,width+left), pandas_options = {"header": 1 })
ds.set_index(ds.columns[0])
```

```python
# taking our value of interest. Now it is the ALL INDIA estimate for 2017
total_production = ds.iloc[-1][1]
total_production
```

```python
total_list = []
for file in file_list:
    try:
        ds = read_pdf(file,pages = 1 ,area = (top,left,height+top,width+left), pandas_options = {"header": 1 })
        ds.set_index(ds.columns[0])
        total_production = ds.iloc[-1][1]
        total_list.append(total_production)
    except:
        print("%s has some problem"%file)
print(total_list)
```

```python
#Total estimate for 2017 is
q = sum(total_list)
q
```

```python
#the last file is the actual production data for 2017
file_list
```

```python
#select the last file and the table of interest is in page 3.
df_actual = read_pdf(file_list[-1],pages = 3 , pandas_options = {"header": 0 })
df_actual.set_index(df_actual.columns[0])
```

```python
p = df_actual.iloc[-1][-1]
diff = p - q
error_estimation = (diff/q)*100
diff
```
